```python
import matplotlib.pyplot as plt
```

```python
Target_Risk=[3,6,9,12,15]
HC=[11.37,11.83,13.82,13.55,12.4]
SM=[6.98,10.69,13.48,14.34,17.19]
GB1=[9.67,13.60,16.03,19.35,21.01]
GB2=[9.77,14.28,16.74,19.71,21.98]
GB3=[10.10,13.32,17.00,24.98,21.01]
SB=[10.01,8.23,5.39,8.82,11.20]
plt.plot(Target_Risk, HC, color='blue', linestyle="-", linewidth="2", markersize="16", marker=".",label="HC")
plt.plot(Target_Risk, SM, color='red', linestyle="-", linewidth="2", markersize="16", marker=".",label="SM")
plt.plot(Target_Risk, GB1 , color='green', linestyle="-", linewidth="2", markersize="16", marker=".",label="GB1")
plt.plot(Target_Risk, GB2 , color='black', linestyle="-", linewidth="2", markersize="16", marker=".",label="GB2")
plt.plot(Target_Risk, GB3 , color='orange', linestyle="-", linewidth="2", markersize="16", marker=".",label="GB3")
plt.plot(Target_Risk, SB , color='pink', linestyle="-", linewidth="2", markersize="16", marker=".",label="SB")
plt.title("Arithmetic mean return")
plt.xlabel("Target_Risk(%)")
plt.ylabel("return(%)")
plt.legend()
plt.show()
```
